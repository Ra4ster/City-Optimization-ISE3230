# Problem Formulation

### Relevant variables

B = $\text{\{school, convenience store, hospital, park, fire station, police station, library, gym, bank, post office\}}$

$$ x^b_{k,l} \in \{0, 1\} \quad\forall (k,l) \in grid\quad\textit{This means each point gets 1 facility type}$$

$x^b_{k,l} = 1$ means facility of type *b* is placed at grid cell *(k, l)*.

$$ y^b_{i,j,k,l} = \begin{cases} 1 \qquad\text{if demand point i,j is assigned to facility of type b located at k,l} \\ 0\qquad \text{otherwise}\end{cases}$$

$y^b_{i,j,k,l} = 1$ means a facility of type *b* serves a customer *(i, j)*, while being located at point *(k, l)*.

**Calculated variables:**

$d_{(i,j),(k,l)} = |i - k| + |j - l|\qquad$ (Manhattan distance)

$w_{i, j} = 1\quad \forall(i,j)\qquad$ (Weights given to points. We assume the population is a constant for all points, IE no less people live in one point than another)

---

### Minimization function: distance to essential buildings

$$\text{min. z} = \sum_{b \in B}\sum_{(i,j)} w_{i,j} \sum_{(k,l)} d_{(i,j),(k,l)} y^b_{i,j,k,l}$$

For our problem with 10 facilities and a 10x10 grid, the problem expands to:

$$\text{min. z} = \sum_{b=1}^{10} \sum_{i=1}^{10} \sum_{j=1}^{10} \sum_{k=1}^{10} \sum_{l=1}^{10} 1(|i-k| + |j-l|)y^b_{i,j,k,l}$$

### Constraints:

$\sum_{(k,l)\in E}y^b_{i,j,k,l} = 1\quad \forall(i,j)\in E,\; \forall b\in B\qquad$ Each demand point is assigned to exactly 1 facility of type b.

$\sum_{b\in B} x^b_{i,j}\leq 1 \quad \forall(i,j)\qquad$ Each point can only hold one facility.

$y^b_{i,j,k,l} \leq x^b_{k,l}\quad\forall(i,j),(k,l)\in E,\;\forall b\in B$ Assignments should only go to existing facilities.

$\sum_{(k,l)\in E} x^b_{k,l} \leq 15 \quad\forall b\in \text{\{hospital, police station, fire station\}}\qquad$ Max 15 hospitals/police/fire stations.

$\sum_{(k,l)\in E} x^b_{k,l} \leq 20 \quad\forall b\in \text{\{convenience, post office, bank\}}\qquad$ Max 20 convenience/post offices/banks.

$\sum_{(k,l)\in E} x^b_{k,l} \leq 10 \quad\forall b\in \text{\{school, gym, library, park\}}\qquad$ Max 10 gyms/libraries/parks.

$\sum_{\substack{(k,\ell)\in E \\ |i-k|+|j-\ell|\le 8}} y^b_{i,j,k,l} = 1\quad\forall b \in \text{\{school, hospital, police station, fire station\}}\qquad$ Max distance 8 to all hospitals/police/fire stations.

$\sum_{\substack{(k,\ell)\in E \\ |i-k|+|j-\ell|\le 20}} y^b_{i,j,k,l} = 1\quad\forall b \in \text{\{convenience, post office, bank\}}\qquad$ Max distance 8 to all convenience/post offices/banks.

$\sum_{\substack{(k,\ell)\in E \\ |i-k|+|j-\ell|\le 15}} y^b_{i,j,k,l} = 1\quad\forall b \in \text{\{gym, library, park\}}\qquad$ Max distance 15 to all gyms/libraries/parks.

```python
m = gp.Model("CityDistance")
m.setParam('OutputFlag', 0)
# VARIABLE PARAMETERS:
grid_size = 10
E = [(i,j) for i in range(1, grid_size+1) for j in range(1, grid_size+1)]
B = ["school", "convenience", "hospital", "park", "fire", "police", "library", "gym", "bank", "post_office"]
# We can edit these statistics as we go:
radius = {
    "school": 8,
    "hospital": 8,
    "police": 8,
    "fire": 8,
    "convenience": 20,
    "post_office": 20,
    "bank": 20,
    "gym": 15,
    "library": 15,
    "park": 15
}
max_facilities = {
    "hospital": 15,
    "police": 15,
    "fire": 15,
    "convenience": 20,
    "post_office": 20,
    "bank": 20,
    "school": 10,
    "gym": 10,
    "library": 10,
    "park": 10
}
```

```python
# Variables:
x = m.addVars(B, range(1,11), range(1,11), vtype=GRB.BINARY, name="x")

y = {}
y = m.addVars(
    [(b,i,j,k,l) for b in B for i,j in E for k,l in E if abs(i-k)+abs(j-l) <= radius[b]],
    vtype=GRB.CONTINUOUS,
    lb=0, ub=1,
    name="y"
)

# Each demand point assigned to exactly one facility of type b
for b in B:
    for i,j in E:
        feasible_y = [y[b,i,j,k,l] for k,l in E if (b,i,j,k,l) in y]
        if feasible_y:
            m.addConstr(gp.quicksum(feasible_y) == 1, name=f"assign_{b}_{i}_{j}")

# Max facilities per type
for b in B:
    m.addConstr(gp.quicksum(x[b,k,l] for k,l in E) <= max_facilities[b], name=f"max_fac_{b}")

# Only one facility per cell
for i,j in E:
    m.addConstr(gp.quicksum(x[b,i,j] for b in B) <= 1, name=f"one_fac_per_cell_{i}_{j}")

M = 100 
# MaxAssignments = len(E) # E is the set of (i,j) coordinates. Assuming len(E) = 100. 
# MaxAssignments = 100 # A clearer name and value.

for b in B:
    for k,l in E:
        # Sum of assignments (i,j) to facility (k,l) of type b
        # Only sum over demand points (i,j) for which the assignment variable y[b,i,j,k,l] exists.
        assignments_sum = gp.quicksum(y[b,i,j,k,l] for i,j in E if (b,i,j,k,l) in y)
        
        # Add the stronger constraint: Sum of assignments to facility (k,l) <= M * x[b,k,l]
        m.addConstr(assignments_sum <= M * x[b,k,l], name=f"strong_couple_{b}_{k}_{l}")

m.update()
# Objective Function:
obj = gp.quicksum(
    (abs(i-k) + abs(j-l)) * y[b,i,j,k,l]
    for b,i,j,k,l in y
)
m.setObjective(obj, GRB.MINIMIZE)
m.optimize()

print("Objective value: z=%f" % m.objVal)
```

```python
for b,k,l in x:
    if x[b,k,l].X > 0.5:
        print("%s placed at (%d, %d)" % (b, k, l))

for b,i,j,k,l in y:
    if y[b,i,j,k,l].X > 0.5:
        print("%s with location (%d, %d) serves location (%d, %d)" % (b,k,l,i,j))
```
