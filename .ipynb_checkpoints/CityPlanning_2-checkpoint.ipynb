# Problem Formulation

### Relevant variables

B = $\text{\{school, convenience store, hospital, park, fire station, police station, library, gym, bank, post office\}}$

$$ x^b_{k,l} \in \{0, 1\} \quad\forall (k,l) \in grid\quad\textit{This means each point gets 1 facility type}$$

$x^b_{k,l} = 1$ means facility of type *b* is placed at grid cell *(k, l)*.

$$ y^b_{i,j,k,l} = \begin{cases} 1 \qquad\text{if demand point i,j is assigned to facility of type b located at k,l} \\ 0\qquad \text{otherwise}\end{cases}$$

$y^b_{i,j,k,l} = 1$ means a facility of type *b* serves a customer *(i, j)*, while being located at point *(k, l)*.

**Calculated variables:**

$d_{(i,j),(k,l)} = |i - k| + |j - l|\qquad$ (Manhattan distance)

$w_{i, j} = 1\quad \forall(i,j)\qquad$ (Weights given to points. We assume the population is a constant for all points, IE no less people live in one point than another)

---

### Minimization function: distance to essential buildings

$$\text{min. z} = \sum_{b \in B}\sum_{(i,j)} w_{i,j} \sum_{(k,l)} d_{(i,j),(k,l)} y^b_{i,j,k,l}$$

For our problem with 10 facilities and a 10x10 grid, the problem expands to:

$$\text{min. z} = \sum_{b=1}^{10} \sum_{i=1}^{10} \sum_{j=1}^{10} \sum_{k=1}^{10} \sum_{l=1}^{10} 1(|i-k| + |j-l|)y^b_{i,j,k,l}$$

### Constraints:

$\sum_{(k,l)\in E}y^b_{i,j,k,l} = 1\quad \forall(i,j)\in E,\; \forall b\in B\qquad$ Each demand point is assigned to exactly 1 facility of type b.

$\sum_{b\in B} x^b_{i,j}\leq 1 \quad \forall(i,j)\qquad$ Each point can only hold one facility.

$y^b_{i,j,k,l} \leq x^b_{k,l}\quad\forall(i,j),(k,l)\in E,\;\forall b\in B$ Assignments should only go to existing facilities.

$\sum_{(k,l)\in E} x^b_{k,l} \leq 25 \quad\forall b\in \text{\{hospital, police station, fire station, convenience store, post office, bank, school, gym, library, park\}}\qquad$ Max 25 hospitals, police stations, fire stations, convenience stores, post offices, banks, schools, gyms, libraries, and  parks.

$\sum_{\substack{(k,\ell)\in E \\ |i-k|+|j-\ell|\le 8}} y^b_{i,j,k,l} = 1\quad\forall b \in \text{\{school, hospital, police station, fire station\}}\qquad$ Max distance 8 to all schools, hospitals, police stations, and fire stations.

$\sum_{\substack{(k,\ell)\in E \\ |i-k|+|j-\ell|\le 10}} y^b_{i,j,k,l} = 1\quad\forall b \in \text{\{convenience store, bank\}}\qquad$ Max distance 10 to all convenience stores and banks.

$\sum_{\substack{(k,\ell)\in E \\ |i-k|+|j-\ell|\le 12}} y^b_{i,j,k,l} = 1\quad\forall b \in \text{\{gym, library\}}\qquad$ Max distance 12 to all gyms and libraries.

$\sum_{\substack{(k,\ell)\in E \\ |i-k|+|j-\ell|\le 15}} y^b_{i,j,k,l} = 1\quad\forall b \in \text{\{post office\}}\qquad$ Max distance 15 to all post offices.

$\sum_{\substack{(k,\ell)\in E \\ |i-k|+|j-\ell|\le 20}} y^b_{i,j,k,l} = 1\quad\forall b \in \text{\{park\}}\qquad$ Max distance 20 to all parks.

```python
m = gp.Model("CityDistance")
m.setParam('OutputFlag', 1)
# VARIABLE PARAMETERS:
grid_size = 10
E = [(i,j) for i in range(1, grid_size+1) for j in range(1, grid_size+1)]
B = ["school", "convenience", "hospital", "park", "fire", "police", "library", "gym", "bank", "post_office"]
# We can edit these statistics as we go:
radius = {
    "school": 8,
    "hospital": 8,
    "police": 8,
    "fire": 8,
    "convenience": 10,
    "bank": 10,
    "gym": 12,
    "library": 12,
    "post_office": 15,
    "park": 20
}
max_facilities = {
    "school": 15,
    "hospital": 15,
    "police": 15,
    "fire": 15,
    "convenience": 15,
    "bank": 15,
    "gym": 15,
    "library": 15,
    "post_office": 15,
    "park": 15
}
```

```python
# OPTIMIZED VERSION
x = m.addVars(B, E, vtype=GRB.BINARY, name="x")

# List of facilities in the coverage of building b at (i,j)
coverage = {}
for b in B:
    for i,j in E:
        coverage[b,i,j] = [(k,l) for k,l in E if abs(i-k)+abs(j-l) <= radius[b]]



# Constraint: One of every b must be placed in the coverage area of each (i,j)
for b in B:
    for i,j in E:
        m.addConstr(gp.quicksum(x[b,k,l] for k,l in coverage[b,i,j]) >= 1)



# Distance between points is **precomputed**:

dist = {}
for b in B:
    for i,j in E:
        for k,l in coverage[b,i,j]:
            dist[b,i,j,k,l] = abs(i-k) + abs(j-l)



# >= 1 building per cell (at most 1 facility per grid cell)
for k,l in E:
    m.addConstr(gp.quicksum(x[b,k,l] for b in B) <= 1, name=f"Max1PerCell_{k}_{l}")

# Objective: Minimize distance to get to all covered buildings:
m.setObjective(
    gp.quicksum(
        dist[b,i,j,k,l] * x[b,k,l]
        for b in B
        for i,j in E
        for k,l in coverage[b,i,j]
    ),
    GRB.MINIMIZE
)

m.optimize()

print(m.objVal)

for var in m.getVars():
    if (var.x == 1):
        print("%s = %d" % (var.varName, var.x))
```

```python
m = gp.Model("NearestFacilityLocation_Corrected")

# x[b, k, l]: Binary variable = 1 if facility type b is placed at (k, l)
x = m.addVars(B, E, vtype=GRB.BINARY, name="x")

# z[b, i, j]: Continuous variable representing the distance from (i, j) to the nearest facility of type b.
z = m.addVars(B, E, vtype=GRB.CONTINUOUS, name="z")

# y[b, i, j, k, l]: Binary variable = 1 if facility (k, l) is CHOSEN as the nearest for demand (i, j)
y_keys = [
    (b, i, j, k, l)
    for b in B
    for i, j in E
    for k, l in coverage[b, i, j]
]
y = m.addVars(y_keys, vtype=GRB.BINARY, name="y")

# =================================================================
# SECTION 3: CONSTRAINTS
# =================================================================

# C1: Coverage Constraint (At least one facility of type b must cover demand point (i, j))
for b in B:
    for i, j in E:
        m.addConstr(gp.quicksum(x[b, k, l] for k, l in coverage[b, i, j]) >= 1,
                    name=f"Coverage_{b}_{i}_{j}")

# C2: Max 1 Per Cell (At most 1 facility of any type per grid cell (k, l))
for k, l in E:
    m.addConstr(gp.quicksum(x[b, k, l] for b in B) <= 1,
                name=f"Max1PerCell_{k}_{l}")

# --- C3, C4, C5: Constraints to define z as the MINIMUM distance ---
# Using a dictionary to structure quicksums for constraints C3 and C4
sum_y = {}
sum_y_dist = {}

for b, i, j, k, l in y.keys():
    # 1. Group the assignment terms by demand point (b, i, j)
    key_b_i_j = (b, i, j)
    if key_b_i_j not in sum_y:
        sum_y[key_b_i_j] = []
        sum_y_dist[key_b_i_j] = []
        
    sum_y[key_b_i_j].append(y[b, i, j, k, l])
    sum_y_dist[key_b_i_j].append(dist[b, i, j, k, l] * y[b, i, j, k, l])
    
    # C5. Linkage Constraint: Can only choose a facility (y=1) that is actually placed (x=1)
    m.addConstr(
        y[b, i, j, k, l] <= x[b, k, l],
        name=f"LinkYtoX_{b}_{i}_{j}_{k}_{l}"
    )

# Now, add the grouped constraints (C3 and C4)
for b, i, j in sum_y.keys():
    # C3. Selection Constraint: Exactly ONE facility must be chosen for assignment (y=1)
    # This choice will be the one that minimizes the total objective.
    m.addConstr(gp.quicksum(sum_y[(b, i, j)]) == 1, name=f"SelectOne_{b}_{i}_{j}")

    # C4. Distance Assignment Constraint: z[b, i, j] is set EQUAL to the distance of the CHOSEN facility
    m.addConstr(z[b, i, j] == gp.quicksum(sum_y_dist[(b, i, j)]), name=f"DistAssignment_{b}_{i}_{j}")


# =================================================================
# SECTION 4: OBJECTIVE AND OPTIMIZE
# =================================================================

# Objective: Minimize the sum of the nearest distances (the total travel distance)
m.setObjective(
    gp.quicksum(z[b, i, j] for b in B for i, j in E),
    GRB.MINIMIZE
)

# Run the optimization
m.optimize()
```

```python
# Plot the 10x10 grid
plt.figure(figsize=(8, 8))
plt.grid(color='black', linestyle='-', linewidth=1.5, zorder=0)
plt.xlim(0.5, 10.5)
plt.ylim(0.5, 10.5)
plt.xticks(range(1, 11))
plt.yticks(range(1, 11))
plt.gca().set_aspect('equal')

# Define color per facility type
facility_colors = {
    'school': 'blue',
    'convenience': 'orange',
    'hospital': 'red',
    'park': 'green',
    'fire': 'darkred',
    'police': 'purple',
    'library': 'brown',
    'gym': 'teal',
    'bank': 'pink',
    'post_office': 'gold'
}

# Plot facilities found in the optimal solution
for b in B:
    for (i, j) in E:
        if x[b, i, j].X > 0.5:   # Facility exists
            plt.scatter(i, j, color=facility_colors[b], s=100, label=b if b not in plt.gca().get_legend_handles_labels()[1] else "")

plt.title("Optimal Facility Locations on City Grid", fontsize=14)
plt.legend(loc='upper right', bbox_to_anchor=(1.3, 1))
plt.show()
```
